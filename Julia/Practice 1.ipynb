### For Loops

```julia
ts_length = 100
epsilon_values = Array(Float64, ts_length) # an empty array for storing 64 bit floating point numbers
for i in 1:ts_length
    epsilon_values[i] = randn()
end
plot(epsilon_values, "b-")
```

### User-Defined Functions

```julia
function generate_data(n)
    epsilon_values = Array(Float64, n)
    for i = 1:n
        epsilon_values[i] = randn()
    end
    return epsilon_values
end

ts_length = 100
data = generate_data(ts_length)
plot(data, "b-")
```

## Exercise 2

```julia
function binomial_rv(n, p)
    ans = rand(n)
    
    for i in 1:n
        if ans[i] < p
            ans[i] = 1
        else
            ans[i] = 0
        end # if
    end # for
    
    return sum(ans)
end #function
```

```julia
binomial_rv(5,1)
```

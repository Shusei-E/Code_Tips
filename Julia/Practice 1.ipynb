### 乱数を発生させてPlotを書く
(Reference for PyPlot)[https://gist.github.com/gizmaa/7214002]

```julia
# Set figure size cf. http://qiita.com/tg-git/items/61b7714665b73b3e4d9a
cm=1/2.54
rc("figure",figsize =([14,7]cm), )
```

```julia
using PyPlot # Here, you can also write `import PyPlot: plot`
ts_length = 150
epsilon_values = randn(ts_length)
plot(epsilon_values, "b-")
```

### For Loops

```julia
ts_length = 100
epsilon_values = Array(Float64, ts_length) # an empty array for storing 64 bit floating point numbers
for i in 1:ts_length
    epsilon_values[i] = randn()
end
plot(epsilon_values, "b-")
```

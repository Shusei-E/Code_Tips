### User-Defined Functions

```julia
function generate_data(n)
    epsilon_values = Array(Float64, n)
    for i = 1:n
        epsilon_values[i] = randn()
    end
    return epsilon_values
end

ts_length = 100
data = generate_data(ts_length)
plot(data, "b-")
```

```julia
function my_factorial(n)
    ans = 1
    for i in 1:n
        ans = ans * i
    end
    return ans
end
```

## Exercise 2

```julia
function binomial_rv(n, p)
    ans = rand(n)
    
    for i in 1:n
        if ans[i] < p
            ans[i] = 1
        else
            ans[i] = 0
        end # if
    end # for
    
    return sum(ans)
end #function
```

```julia
binomial_rv(5,1)
```

## Exercise 3

```julia
n = 100000000

count = 0
for i in 1:n
    u, v = rand(2) # A point in the square
    d = sqrt((u - 0.5)^2 + (v - 0.5)^2)  # Distance from middle of square
    if d <=0.5
        count += 1
    end
end

area_estimate = count / n

print(area_estimate / (1/4))  # dividing by radius**2
```

## Exercise 4

```julia
count = 0
for i in 1:10
    if rand() > 0.5
        count += 1
        
        if count == 3
            print("Pay")
        end # if
    else
        count = 0
    end # if
end # for
```

#### ternary operator

```julia
payoff = 0
count = 0

print("Count = ")

for i in 1:10
    U = rand()
    count = U < 0.5 ? count + 1 : 0  # ternary operator is used
    print(count)
    if count == 3
        payoff = 1
    end
end
print("\n")
println("payoff = $payoff")
```

##### Another Answer

```julia
alpha = 0.9
T = 200
x = zeros(T + 1)

for t in 1:T
    x[t+1] = alpha * x[t] + randn()
end
plot(x, color="blue")
```

## Exercise 6

```julia
alphas = [0.0, 0.8, 0.98]
T = 200

series = []
labels = []

for alpha in alphas
    x = zeros(T + 1)
    x[1] = 0
    for t in 1:T
        x[t+1] = alpha * x[t] + randn()
    end
    push!(series, x)
    push!(labels, "alpha = $alpha")
end

for i in 1:3
    plot(series[i], label  = labels[i])
end
legend(loc=3, fontsize=7)
title("Exercise 6")
```
